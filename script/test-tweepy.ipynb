Hay dos formas de obtener la localizaci√≥n de un tweet:

* el usuario puede a√±adirla manualmente (geotagging). S√≥lo un 0,85% de los tweets llevan esta informaci√≥n, por lo que no me es √∫til: necesitar√≠a analizar muchos m√°s tweets de los que me permite la API para obtener suficientes datos
* utilizar la localizaci√≥n de la cuenta del usuario (user.location)

Tweepy tiene una funci√≥n "flatten" que agrupa los contenidos del Paginator en un generador, pero no puedo usarla porque no almacena la informaci√≥n de localizaci√≥n.

Hacemos que cada usuario s√≥lo sea analizado una vez (al d√≠a).

```python
tweetList = []
listaUsuarios = []

# Cuidado: limit es el n√∫mero de p√°ginas, no de tweets!!
# si queremos 1000 tweets, limit = 10

for response in tweepy.Paginator(client.search_recent_tweets, 
query=query, max_results=100, start_time=currentDay, expansions= 'author_id', user_fields=['location'], limit=2):

    # S√≥lo quiero analizar un m√°ximo de un tweet por persona
    # Si un usuario ha escrito varios tweets, aparece varias veces en la lista de tweets pero s√≥lo una
    # en la lista de usuarios, por lo que necesito un contador separado para esta lista
    contadorUsuarios = 0

    for i in range(response.meta['result_count']):

        autor = response.data[i]['author_id']
        if autor in listaUsuarios:            
            continue
        listaUsuarios.append(autor)
        tweetList.append({'id': response.data[i].id, 'text': response.data[i].text, 
        'location': response.includes['users'][contadorUsuarios].location})
        contadorUsuarios += 1
```
